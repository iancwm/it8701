# Theme 1 - Telecommunications service standards

We take a look at how well covered Singapore is by its main telecommunications providers, and how reliable their service is. The data consists of a timeseries of percentages representing coverage and drop rates for 3G services.

## Datasets used

1. **Dataset 1** - `3g-public-cellular-mobile-telephone-services-nation-wide-service-coverage.csv`
2. **Dataset 2** - `3g-public-cellular-mobile-telephone-services-average-drop-call-rate.csv`

## Methodology
The data sets are one variable time series data with categories.

We use lineplots on both datasets and a boxplot to determine the variability in performance.

## Set up
Load the data using `pandas.read_csv` and print the `info()` to get a sense of the data

```python
df1 = pd.read_csv(data1)
df2 = pd.read_csv(data2)
for i in (df1,df2):
    print(i.info())
```

**Dataset 1**

We use `seaborn` to generate a `lineplot`, split by `telco`.

```python
fig,ax=plt.subplots()
sns.lineplot(data=df1,
             x='month',
             y='service_coverage',
             hue='telco').set_title('Singapore Telcos Nationwide Service Coverage (%)')
plt.xlabel('Month')
plt.ylabel('Service Coverage (%)')
plt.xticks(rotation='vertical')
month=df1['month'].iloc[len(df1)-1]
most_recent=df1[df1['month']==month]
for i in range(len(most_recent)):
    ax.text(month, 
            most_recent['service_coverage'].iloc[i], 
            "{}%".format(str(most_recent['service_coverage'].iloc[i])), 
            color=current_palette[i], 
            ha='left')
plt.show()
```

We can plot a `boxplot` to visualize the series' distribution

```python
sns.boxplot(data=df1,
            x='telco',
            y='service_coverage',
            notch=True).set_title('Singapore Telcos Nationwide Service Coverage Boxplot')
plt.xlabel('Telco Provider')
plt.ylabel('Service Coverage (%)')
plt.show()
```

**Dataset 2**

We give the same treatment to Dataset 2, calling `seaborn` to generate a `lineplot` by `telco`.

```python
fig, ax = plt.subplots()
sns.lineplot(data=df2,
             x='month',
             y='drop_call_rate',
             hue='telco').set_title('Singapore Telco Average 3G Call Drop Rate (%)')
plt.xlabel('Month')
plt.ylabel('Call Drop Rate (%)')
plt.xticks(rotation='vertical')

# Labels
month=df2['month'].iloc[len(df2)-1]
most_recent=df2[df2['month']==month]
for i in range(len(most_recent)):
    ax.text(month, 
            most_recent['drop_call_rate'].iloc[i], 
            "{}%".format(str(most_recent['drop_call_rate'].iloc[i])), 
            color=current_palette[i], 
            ha='left')
plt.show()
# change tick size
```

Generating `boxplot` for the three telcos.

```python
sns.boxplot(data=df2,x='telco',y='drop_call_rate').set_title('Singapore Telco Average 3G Call Drop Rate Boxplot')
plt.xlabel('Telco Provider')
plt.ylabel('Drop Call Rate (%)')
plt.show()
```

```python
df2_pie = df2[-3:].set_index('telco')['drop_call_rate']
df2_pie_percentage = df2_pie / df2_pie.sum()
df2_pie_percentage.plot(kind='pie',
                        autopct='%1.1f%%',
                        textprops={'color':'white',
                                   'weight':'bold'}).set_title('June 2016 3G Call Drop Share by Telco')
plt.show()
```

## Analysis

In terms of island-wide coverage (Dataset 1), it would seem M1 is the most erratic from a visual comparison of the boxplot we generated, and we may think it justified to deem it the worst of the three telcos.

However, bear in mind on an absolute scale, all three major telcos are doing a good job, with coverage well-above 95% across the board.

Furthermore, from Dataset 2, while we can see that the telcos have lowered the drop call rate on average, we can see that Singtel is responsible for most of 3G call drops as of June 2016, with its `drop_call_rate` the highest of the three.

So, while Singtel wins in terms of sheer coverage, the service is marginally less reliable.

# Theme 2 - Singapore mobile broadband usage trends

*How much* of the population has **access** to mobile data, and how does this compare to the rest of the world?

Also, just *how much* mobile data is Singapore **consuming**?

## Datasets used

3. **Dataset 3** - `mobile-data-usage.csv`
4. **Dataset 4** - `mobile-penetration-rate.csv`
5. **World Bank Dataset 1** - `API_IT.CEL.SETS.P2_DS2_en_csv_v2_54223`

## Methodology
Simple one variable time-series data warrants a lineplot.

## Set up
Load data with `pandas.read_csv`

```python
df3 = pd.read_csv(data3)
df4 = pd.read_csv(data4)
worldbankdf1 = pd.read_csv(worldbankdata1,skiprows=4,header=0)
for i in (df3,df4,worldbankdf1):
    print(i.info())
```

### Dataset 3
Generate a bar chart showing Mobile Data consumption, complete with labels.

```python
fig,ax=plt.subplots()
sns.barplot(data=df3[14:], # select more recent time periods
            x='quarter',
            y='volume_of_mobile_data',
            color=current_palette[0],
            edgecolor='None').set_title('Singapore Volume of Mobile Data (PB)')
# Plot value labels 0.5 units above the height of each bar
for i, v in enumerate(df3['volume_of_mobile_data'][14:].values.round(1)):
    ax.text(i, 
            v+0.5, 
            str(v), 
            color=current_palette[0], 
            ha='center')
plt.xlabel('Quarter')
plt.ylabel('Mobile Data Usage (PB)')
plt.xticks(rotation='vertical')
plt.show()
# change tick size
```

Then, we use the `np.diff` function to calculate percentage change each quarter for the data, and plot the resulting line plot

```python
fig,ax=plt.subplots()
df3_growth = pd.DataFrame(df3['volume_of_mobile_data'].diff()/df3['volume_of_mobile_data'][1:]*100).set_index(df3['quarter'])
sns.lineplot(data=df3_growth).set_title('% Growth in Mobile Data Usage (Quarterly)')

# Labels
month=df3_growth.index[-1]
ax.text(month,
        df3_growth.iloc[len(df3_growth)-1]['volume_of_mobile_data'],
        str("{}%".format(df3_growth.iloc[len(df3_growth)-1]['volume_of_mobile_data'].round(1))),
        color=current_palette[0],
        fontsize='large',
        ha='left')
plt.xlabel('Quarter')
plt.ylabel('Growth in Mobile Data Usage (%)')
plt.xticks(rotation='vertical')
plt.show()
```

### Dataset 4
We generate a `lineplot` to analyze trends in mobile penetration in Singapore. We filter out earlier time periods to leave sufficient space on the x-axis

```python
plt.figure(figsize=(15,6))
fig,ax=plt.subplots()
sns.lineplot(data=df4[100:], # Filter out older dates
             x='month',
             y='mobile_penetration_rate').set_title('Singapore mobile penetration (%)')
# Labels
month=df4['month'].iloc[-1]
ax.text(month,
        df4.iloc[len(df4)-1]['mobile_penetration_rate'],
        str("{}%".format(df4.iloc[len(df4)-1]['mobile_penetration_rate'].round(1))),
        color=current_palette[0],
        fontsize='large',
        ha='left')
plt.xlabel('Month')
plt.ylabel('Mobile Penetration (%)')
plt.xticks(rotation='vertical',fontsize=4)
plt.show()
```

How is Singapore faring in relation to the world in terms of mobile peneration? 

We take a look at the dataset from **World Bank Data**

```python
worldbankdf1.info()
```

The dataset is not in the shape we want it, as the series are the years, and the countries are the values.

We need to drop irrelevant labels, transpose the set, and set the new columns and indexes.

```python
# Transpose the dataset, set column labels to the original index which is now in the dataframe, and then drop them
worldbankdf1_cleaned = worldbankdf1.drop(labels=['Country Code','Indicator Name','Indicator Code'],
                                         axis=1).T
worldbankdf1_cleaned.columns = worldbankdf1_cleaned.iloc[0]
# The dataframe is now made of objects, so we need to cast them to numerical data for analysis
worldbankdf1_cleaned = worldbankdf1_cleaned.drop(['Country Name','Unnamed: 63']).ffill().astype('float64')
```

```python
worldbankdf1_cleaned
```

Let's generate a bar chart comparing Singapore's mobile penetration rate to G10 economies.

```python
mobile_penetration_g10_singapore = worldbankdf1_cleaned.loc['2017'][['Singapore',
                                                                     'Belgium',
                                                                     'Canada',
                                                                     'France',
                                                                     'Italy',
                                                                     'Japan',
                                                                     'Netherlands',
                                                                     'Sweden',
                                                                     'Switzerland',
                                                                     'United Kingdom',
                                                                     'United States']].sort_values()

fig, ax = plt.subplots()
sns.barplot(x=mobile_penetration_g10_singapore.index,
            y=mobile_penetration_g10_singapore,
            palette=current_palette_ext,
            edgecolor='None').set_title('Singapore has higher mobile penetration than the G10 economies (2017)')
# Draw labels with y-offset of 3, and colors iterated according to colour of the series
for i, v in enumerate(mobile_penetration_g10_singapore.values.round(1)):
    ax.text(i, 
            v+3, 
            str(v), 
            color=current_palette_ext[i], 
            ha='center')
plt.ylabel('Mobile Penetration %')
plt.xlabel('Country')
plt.xticks(rotation='vertical')
plt.show()
```

## Analysis

From the data we have, its plain to see Singapore residents are very much 'plugged in'. 

Mobile penetration stands at over 100%, which suggests people own multiple devices that are connected to the internet. 

Singapore's mobile penetration surpasses even the G10 countries.

Mobile data consumption is also growing exponentially, a sign of an increasingly interconnected society and comfort with technology.

### Dataset 7
This dataset is non-trivial. Lets use `DataFrame.info()` to detect `NaN` values.

```python
df7 = df7.replace('na',np.NaN)
df7.info()
```

**Observations:** There are 2433 entries but `price_of_plan` and `price_per_mbps` have far fewer non-null values, indicating significant presence of NaN values.

Also, the `DataFrame` is comprised of objects. This includes the numerical fields and the date columns. We need to remedy the data types to facilitate analysis

```python
df7['max_speed']=pd.to_numeric(df7['max_speed'])
df7['price_of_plan']=pd.to_numeric(df7['price_of_plan'])
df7['price_per_mbps']=pd.to_numeric(df7['price_per_mbps'])
df7['date_of_retrieval_'] = pd.to_datetime(df7['date_of_retrieval_'],
                                           dayfirst=True,
                                           format='%d/%m/%Y')
df7['month'] = pd.to_datetime(df7['month'],
                              format='%Y-%m')
```

```python
df7[df7['price_per_mbps'].isnull()][1500:]
```

```python
plan_average = df7.groupby('plan').mean()
for i in plan_average.index:
    try:
        df7['price_of_plan'].loc[df7['price_of_plan'].isnull() & df7['plan']==i] = plan_average.loc[i]['price_of_plan']
        df7['price_per_mbps'].loc[df7['price_per_mbps'].isnull() & df7['plan']==i] = plan_average.loc[i]['price_per_mbps']
        df7['max_speed'].loc[df7['max_speed'].isnull() & df7['plan']==i] = plan_average.loc[i]['max_speed']
    except:
        print('Plan {} has no values to fill'.format(i))
df7
```

The dataset still has many `NaN` values. We will refrain from filling in with broad averages since this will distort the data too much.

Since there is a wide range of data, we use `plotly_express` to create an interactive chart. Feel free to lasso the area you wish to analyze deeper, and double click to return to the original view.

We plot a jointplot to combine a `scatterplot` and `rugplots` for the y-axis as well as `histograms` for the x-axis

```python
fig = px.scatter(df7, 
                 x="max_speed", 
                 y="price_of_plan", 
                 color="operator",
                 color_discrete_sequence=px.colors.colorbrewer.Set2,
                 hover_name='month',
                 template='seaborn',
                 marginal_y='histogram',
                 marginal_x='rug')
fig.show()
```

In order to viusalize the distribution of plans by value for money, we plot a cumulative frequency distribution of the pan SGD/mbps.

```python
sns.distplot(df7['price_per_mbps'].dropna(),
             bins=20,
             hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True)).set_title('Cumulative Frequency of Plans by SGD/Mbps')
plt.xlabel('SGD/Mbps')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(data=df7,
            x='operator',
            y='price_per_mbps',
            notch=True).set_title('Singapore Telco Pricing Boxplot')
plt.xlabel('Telco Operator')
plt.ylabel('SGD/Mbps')
plt.show()
```

```python
sns.boxplot(data=df7,
            x='connection_type',
            y='price_per_mbps',
            notch=True).set_title('Singapore Telco Pricing Boxplot')
plt.xlabel('Contract Type')
plt.ylabel('SGD/Mbps')
plt.show()
```

At first glance, it would seem that Singtel is lagging behind the pack in terms of value of money.

After all, it has many offerings that cost significantly higher per MBPS than its competitors.

However, we note that ADSL connections are signficantly more expensive than Fibre and Fixed Broadband, which tend be more common nowadays.

ADSL is a legacy offering that is technologically inferior to Fibre and Fixed Broadband.

Let's analyze telco offerings again, less ADSL.

```python
sns.boxplot(data=df7[df7['connection_type'] !='ADSL'],
            x='operator',
            y='price_per_mbps',
            notch=True).set_title('Singapore Telco Pricing Boxplot (less ADSL)')
```

Controlling for ADSL, we actually find that Singnet offers better value than the other two major telcos. 

Hence, Singnet's legacy offerings of ADSL services was skewing the data.

```python
df7[df7['connection_type'] != 'ADSL'].describe()
```

## Analysis

At an average of SGD 0.38 per mbps per month, Singapore telcos offer good value for money on a global scale, beating even the G10.

# MySQL
We use `sqlalchemy` with `mysql.connector` to make a connection to a database `ca2ian` and upload Dataset 7 for later retrieval.

```python
import mysql.connector
from sqlalchemy import create_engine, types
user, pw, host, port, db , table= 'root','(Hl2freak)','127.0.0.1','3306','ca2ian','singapore_broadband_prices'
engine = create_engine('mysql+mysqlconnector://' + user + ':' + pw + '@' + host + ':' + port + '/' + db , echo=False)
```

We specify the `dtype` of the columns in a list, then `zip` it with the columns to produce a `dict` mapping column names to `dtypes`.

```python
df7_sql_dtype=[types.Date(),
               types.Date(),
               types.VARCHAR(30),
               types.VARCHAR(60),
               types.SMALLINT(),
               types.VARCHAR(16),
               types.VARCHAR(16),
               types.VARCHAR(16),
               types.SMALLINT(),
               types.SMALLINT()]
df7_dtypes=dict(zip(df7.columns,df7_sql_dtype))
for k, v in df7_dtypes.items():
    print(k,"\t",v)
```

```python
df7.to_sql(name=table,
           con=engine,
           if_exists='replace',
           index=True,
           dtype=df7_dtypes)
```

```python
sql='SELECT operator, connection_type, AVG(max_speed), AVG(price_of_plan) FROM {} WHERE (max_speed IS NOT NULL) & (price_of_plan IS NOT NULL) GROUP BY operator'.format(table)
df_sql= pd.read_sql_query(sql=sql,
                          con=engine)
print(df_sql)
```

```python
px.bar(df_sql,
       x='operator',
       y='AVG(max_speed)',
       color='operator',
       color_discrete_sequence=px.colors.colorbrewer.Set2)
```

```python
px.bar(df_sql,
       x='operator',
       y='AVG(price_of_plan)',
       color='operator',
       color_discrete_sequence=px.colors.colorbrewer.Spectral)
```
